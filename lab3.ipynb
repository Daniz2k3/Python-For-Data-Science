```python
Q1 = df['mass'].quantile(0.25)
Q3 = df['mass'].quantile(0.75)

filter_df = df[(df['mass'] >= Q1) & (df['mass'] <= Q3)]

# Xuat ra cac dong co mass nam trong khoang Q1 va Q3
filter_df
```

```python
max_min_fruit=df.groupby('fruit_name')['color_score'].agg(['max','min'])
print(max_min_fruit)
```

```python
mean_std_fruit=df.groupby('fruit_name')['color_score'].agg(['mean','std'])
print(mean_std_fruit)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['fruit_subtype'] = le.fit_transform(df['fruit_subtype'])

df_2 = df[['mass','width','height','color_score','fruit_subtype','fruit_name']]
df_2
```

# Phan 2

```python
df_2['fruit_subtype'] = df_2['fruit_subtype'].astype('category')
df_2['fruit_name'] = df_2['fruit_name'].astype('category')
df_2.dtypes
```

```python
new_df = df_2[(df_2['fruit_name'] == "apple") | (df_2['fruit_name'] == "lemon")]
new_df
```

```python
from sklearn.model_selection import train_test_split

X = new_df.drop('fruit_name', axis=1)
y = new_df['fruit_name']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Chon model GaussianNB vi phu hop voi gia dinh phan phoi chuan cho cac dac trung (mass, width, height, color_socre)

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB(priors=None, var_smoothing=1e-02)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
print(f"{accuracy_score(y_test, y_pred):.4f}")
```

```python
X = df_2.drop('fruit_name', axis=1)
y = df_2['fruit_name']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB

MNB = MultinomialNB(alpha=0, fit_prior=False, class_prior=None)

MNB.fit(X_train, y_train)

y_pred = MNB.predict(X_test)
```

```python
print(f"{accuracy_score(y_test, y_pred):.4f}")
```

```python
report = classification_report(y_test, y_pred, output_dict=True)

# Extract and print the accuracy for each fruit
for fruit, metrics in report.items():
    if fruit in y_test.cat.categories:
        print(f"Accuracy for {fruit}: {metrics['precision']:.4f}")
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

import matplotlib.pyplot as plt

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred, labels=y_test.cat.categories)

# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=y_test.cat.categories, yticklabels=y_test.cat.categories)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
