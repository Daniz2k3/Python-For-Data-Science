```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./Dataset/Fruit Dataset.txt", sep="\t")
df.head(10)
```

```python
print(f"So dong {df.shape[0]}, so cot {df.shape[1]}")
```

```python
column_list = df.columns.tolist()
print(column_list)
```

```python
describe_df = df.describe()
describe_df
```

```python
# Danh sách các cột cần tìm min và max
columns = ['mass', 'width', 'height', 'color_score']

# In ra dòng chứa giá trị max của các cột
print("\tDòng chứa giá trị max của các cột:")
for col in columns:
    print(f"{col}:")
    print(df[df[col] == df[col].max()])

# In ra dòng chứa giá trị min của các cột
print("\n\tDòng chứa giá trị min của các cột:")
for col in columns:
    print(f"{col}:")
    print(df[df[col] == df[col].min()])
```

```python
Q1 = df['mass'].quantile(0.25)
Q3 = df['mass'].quantile(0.75)

filter_df = df[(df['mass'] >= Q1) & (df['mass'] <= Q3)]

# Xuat ra cac dong co mass nam trong khoang Q1 va Q3
filter_df
```

```python
max_min_fruit=df.groupby('fruit_name')['color_score'].agg(['max','min'])
print(max_min_fruit)
```

```python
mean_std_fruit=df.groupby('fruit_name')['color_score'].agg(['mean','std'])
print(mean_std_fruit)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['fruit_subtype'] = le.fit_transform(df['fruit_subtype'])

df_2 = df[['mass','width','height','color_score','fruit_subtype','fruit_name']]
df_2
```

# Phan 2

```python
new_df = df_2[(df_2['fruit_name'] == "apple") | (df_2['fruit_name'] == "lemon")]
print(new_df)
```

```python
from sklearn.model_selection import train_test_split

X = new_df.drop('fruit_name', axis=1)
y = new_df['fruit_name']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

Chon model GaussianNB vi phu hop voi gia dinh phan phoi chuan cho cac dac trung (mass, width, height, color_socre)

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB(priors=None, var_smoothing=1e-08)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
print(model.score(X_test, y_test))
```
