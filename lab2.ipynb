## 1. Xac dinh van de: Tinh xac suat de co it nhat 2 ung vien dau phong van

De bai cho:
- Xac suat dau phong van cua ung vien A: $P(A) = 0.6$
- Xac suat dau phong van cua ung vien B: $P(B) = 0.1$
- Xac suat dau phong van cua ung vien C: $P(C) = 0.2$

### Step 1: Tinh xac suat that bai

Goi $A'$, $B'$, va $C'$ lan luot la xac suat de 3 ung vien A, B, and C truot phong van:

- $P(A') = 1 - P(A) = 1 - 0.6 = 0.4$
 
- $P(B') = 1 - P(B) = 1 - 0.1 = 0.9$

- $P(C') = 1 - P(C) = 1 - 0.2 = 0.8$


### Step 2: Tinh xac xuat co it hon 2 ung vien dau phong van:
- **Khong co ung vien nao dau**:

$$
    P(\text{khong ai dau}) = P(A') \times P(B') \times P(C') = 0.4 \times 0.9 \times 0.8 = 0.288
$$

- **Co duy nhat mot ung vien dau**:

$$
\begin{align}
P(\text{mot nguoi}) &= P(A) \times P(B') \times P(C') + P(A') \times P(B) \times P(C') + P(A') \times P(B') \times P(C) \\
&= (0.6 \times 0.9 \times 0.8) + (0.4 \times 0.1 \times 0.8) + (0.4 \times 0.9 \times 0.2) \\
&= 0.432 + 0.032 + 0.072 \\
&= 0.536
\end{align}
$$

## Step 3: Tinh xac suat de co it nhat hai nguoi dau phong van:
Tong xac suat de co it hon 2 nguoi dau phong van:

$$
P(\text{it hon 2 nguoi}) = P(\text{khong ai dau}) + P(\text{mot nguoi}) = 0.288 + 0.536 = 0.824
$$

Vi vay, xac suat de co it nhat 2 nguoi dau phong van la:

$$
P(\text{it nhat 2 nguoi}) = 1 - P(\text{it hon 2 nguoi}) = 1 - 0.824 = 0.176
$$

## Ket qua cuoi cung

Xac suat de co it nhat 2 nguoi dau phong van la: 0.176 hay 17.6%.

## 2. Tính xác suất để rút được 5 lá rô, 2 lá cơ, và 2 lá K trong bộ 52 lá

- **Tổng số cách chọn 9 lá từ bộ bài 52 lá**:
  $$
  \Omega = \binom{52}{9}
  $$

- **Th1: có 1 lá già cơ và 1 lá già rô**: 

  số cách chọn:
  $
  A = \binom{12}{4} \cdot \binom{12}{1} \cdot \binom{26}{2}
  $

- **Th2: có 1 già cơ:**:

  số cách chọn:
  $
  B = \binom{2}{1} \cdot \binom{13}{5} \cdot \binom{12}{1} \cdot \binom{26}{1}
  $

- **Th3: có 1 lá già rô:**:

  có số cách chọn:
  $
  C = \binom{2}{1} \cdot \binom{12}{4} \cdot \binom{13}{2} \cdot \binom{26}{1}
  $

- **Th: chỉ có già chuồn và già bích:**:

  có số cách chọn:
  $
  D = \binom{13}{5} \cdot \binom{13}{2}
  $

## Vậy xác suất để rút được 5 lá rô, 2 lá cơ và 2 lá K trong bộ 52 lá là:

$$
\begin{align}
P(\text{X}) = \frac{A = B + C + D}{\binom{52}{9}}
&\approx 1.316 \times 10^{-3} = 0.1316 \%
\end{align}
$$


## 3. Tim ngan hang co kha nang bi gian lan nhat trong truong hop bi phat hien

#### Theo de bai ta co:

- xac suat giao dich den tu ngan hang A: $P(A) = 0.3$ va Xac suat gian lan cua ngan hang A: $P(Fraud|A) = 0.03$

- xac suat giao dich den tu ngan hang B: $P(A) = 0.2$ va Xac suat gian lan cua ngan hang B: $P(Fraud|B) = 0.05$

- xac suat giao dich den tu ngan hang C: $P(A) = 0.5$ va Xac suat gian lan cua ngan hang C: $P(Fraud|A) = 0.04$

#### Ap dung cong thuc xac suat toan phan:

$$
\begin{align}
P(Fraud) &= P(A) \times P(Fraud|A) + P(B) \times P(Fraud|B) + P(C) \times P(Fraud|C) \\
&= 0.3 \times 0.03 + 0.2 \times 0.05 + 0.5 \times 0.04 \\
&= 0.039
\end{align}
$$

#### Ap dung cong thuc Bayes:

$P(A|Fraud) = \frac{P(Fraud|A) \times P(A)}{P(Fraud)} = \frac{0.03 \times 0.3}{0.039} \approx 0.231 \approx 23.1\%$

$P(B|Fraud) = \frac{P(Fraud|B) \times P(B)}{P(Fraud)} = \frac{0.05 \times 0.2}{0.039} \approx 0.256 \approx 25.6\%$

$P(C|Fraud) = \frac{P(Fraud|C) \times P(C)}{P(Fraud)} = \frac{0.04 \times 0.5}{0.039} \approx 0.513 \approx 51.3\%$

### $\Rightarrow$ Vay ngan hang C co nhieu kha nang nhat(51.3%) neu giao dich bi phat hien gian lan.

## 4. Xay dung khoang tin cay cho cho trung binh voi do tin cay 90%
#### Ta co:
- n = 250000 mau ngau nhien
  
- Trung binh $\bar{X}$ = 525000 VND
  
- Do lech chuan $\sigma$ = 28340

- Phan vi tren muc ${\alpha/2}$  cua T: $t_{\alpha/2}$ = 1.711

- ${\alpha}$ = 1 - 0.9 = 0.1

#### Khoang tin cay voi muc tin cay 90%:

$$
\begin{align}
\bar{X} - t_{\alpha/2} \times \frac{\sigma}{\sqrt{n}} \leq \mu \leq \bar{X} + t_{\alpha/2} \times \frac{\sigma}{\sqrt{n}} \\
\Leftrightarrow 525000 - 1.711 \times \frac{28340}{\sqrt{250000}} \leq \mu \leq 525000 + 1.711 \times \frac{28340}{\sqrt{250000}} \\
\Rightarrow \text{Khoang tin cay} \approx (524903 , 525097)
\end{align}
$$

# 5. recommendation system

```python
df
```

Dung ky thuat **"Item-based Collaborative Filtering"** de dung lich su mua hang cua tung **UserId** Tu do de xuat cac san pham tuong tu dua vao khoang cach **Cosine Similarity** tinh duoc.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Tao ma tran nguoi dung - san pham
user_item_matrix = df.pivot_table(index='UserId', columns='ItemCode', values='NumberOfItemsPurchased', fill_value=0)

# Tinh do tuong dong cosine giua cac san pham
item_similarity = cosine_similarity(user_item_matrix.T)

# Ham goi y san pham
def recommend_items(user_id, num_recommendations=5):
    if user_id not in user_item_matrix.index:
        return f"User ID {user_id} not found."
    user_purchases = user_item_matrix.loc[user_id]
    scores = pd.Series(item_similarity.dot(user_purchases), index=user_item_matrix.columns)
    top_items = scores.nlargest(num_recommendations + len(user_purchases))
    return top_items[~top_items.index.isin(user_purchases.index[user_purchases > 0])].index.tolist()

```

```python
sp = recommend_items(384027)

print("Top 5 san pham ma nguoi dung co ID: 384027 co the quan tam:")
for i in range(10):
    print(df['ItemDescription'][df['ItemCode'] == sp[i]].unique())
```
